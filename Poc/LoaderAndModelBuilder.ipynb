```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
import mlflow
import mlflow.sklearn
from urllib.parse import urlparse
```

```python
def create_dataframe(subset_type: str)-> pd.DataFrame:
    """
    See https://scikit-learn.org/0.19/datasets/twenty_newsgroups.html
    """
    newsgroups_data = fetch_20newsgroups(subset=subset_type)
    data = {
        'text': newsgroups_data.data,
        'category_number' : newsgroups_data.target
    }

    df = pd.DataFrame(data)
    df['category'] = df['category_number'].apply(lambda x: newsgroups_data.target_names[x])  
    
    return df
```

```python
df_train = create_dataframe('train')
df_train.head()
```

### Example Data Cleaning
Take only articles from 5 categories:

- `rec.sport.hockey`
- `rec.autos`
- `sci.med`
- `sci.electronics`
- `talk.politics.misc`

```python
from typing import List
def select_only_categories(allowed_categories: List[str], df: pd.DataFrame) -> pd.DataFrame:
    df = df.loc[df['category'].isin(allowed_categories)]
    return df
```

```python
selected_columns = ['rec.sport.hockey', 'rec.autos', 'sci.med', 'sci.electronics', 'talk.politics.misc']
df_train = select_only_categories(selected_columns, df_train)
df_test = create_dataframe('test')
df_test = select_only_categories(['rec.sport.hockey', 'rec.autos', 'sci.med', 'sci.electronics', 'talk.politics.misc'], df_test)
df_train['category'].value_counts()
```

## Build Pipeline & Log Model
https://databricks.com/notebooks/reprisk_notebooks/03_rep_modelling.html

- Register a Model: https://mlflow.org/docs/latest/model-registry.html#adding-an-mlflow-model-to-the-model-registry

```python
def store_confusion_matrix(predicted, df_test, run_id):
    # create confusion matrix
    pred_df = pd.DataFrame(zip(predicted, df_test['category']), columns=['predicted', 'actual'])
    confusion_matrix = pd.crosstab(pred_df['actual'], pred_df['predicted'], rownames=['label'], colnames=['prediction'])

    # plot confusion matrix
    plt.figure(figsize=(6,5))
    sns.heatmap(confusion_matrix, annot=True, cmap="Blues", fmt='d')
    plt.savefig("confusion_matrix.png")

    # store confusion matrix alongside model on mlflow
    client = mlflow.tracking.MlflowClient()
    client.log_artifact(run_id, "confusion_matrix.png")
    plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
from mlflow.tracking import MlflowClient

model_name = "mlflow_poc"
# If this database does not exist, mlflow will create it 
mlflow.set_tracking_uri("sqlite:///mlflow_database.db")

mlflowclient = MlflowClient(
    mlflow.get_tracking_uri(), mlflow.get_registry_uri())

mlflow.set_experiment("MLFLOW_TEST")
with mlflow.start_run(run_name=model_name):
    alpha = 0.5
    run_id = mlflow.active_run().info.run_id
    
    text_clf = Pipeline([
        ('tfidf', TfidfVectorizer()),
        ('clf', MultinomialNB(alpha)),
    ])
    
    
    mlflow.log_param("alpha", alpha)
    text_clf.fit(df_train['text'], df_train['category'])
    predicted = text_clf.predict(df_test['text'])
    
    accuracy = metrics.accuracy_score(predicted, df_test['category'])
    
    # Register model
    tracking_url_type_store = urlparse(mlflow.get_tracking_uri()).scheme
    #print(f"{tracking_url_type_store=}")

    # Model registry does not work with file store
    if tracking_url_type_store != "file":
        mlflow.sklearn.log_model(
            text_clf, "naive-bayes", registered_model_name=model_name, conda_env=mlflow.sklearn.get_default_conda_env())
    else:
        mlflow.sklearn.log_model(
            text_clf, "naive-bayes", registered_model_name=model_name)
        
    # Transition to production. We search for the last model with the name and we stage it to production
    mv = mlflowclient.search_model_versions(
        f"name='{model_name}'")[-1]  # Take last model version
    mlflowclient.transition_model_version_stage(
        name=mv.name, version=mv.version, stage="production")

    
    mlflow.log_metric("accuracy", accuracy)

    
    # Log classification report
    clsf_report = pd.DataFrame(metrics.classification_report(predicted, df_test['category'], output_dict=True)).transpose()
    clsf_report.to_csv('classification_report.csv', index=True)
    mlflow.log_artifact('classification_report.csv')
    
    print(metrics.classification_report(df_test['category'], predicted))
    
    store_confusion_matrix(predicted, df_test, run_id)
    

    plt.show()
```

## Fetch Model
- If you use mlflow.sklearn.log_model(text_clf, "pipeline") to save the model than you have to use loaded_model = mlflow.pyfunc.load_model(logged_model)
- If you use `    mlflow.sklearn.log_model(
        sk_model=text_clf,
        artifact_path="sklearn-model",
        registered_model_name=model_name
    )` then we need to use model = mlflow.pyfunc.load_model(
    model_uri=f"models:/{model_name}/{model_version}"
)



```python
import mlflow.pyfunc

#model_name = "sk-learn-random-forest-reg-model"
model_version = 1

model = mlflow.pyfunc.load_model(
    model_uri=f"models:/{model_name}/{model_version}"
)

df_test['predicted_from_loaded'] = model.predict(df_test['text'])
```
